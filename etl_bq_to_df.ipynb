```python
from google.cloud import bigquery
client = bigquery.Client()
project = "ugm-yt-etl"
dataset_id = "youtube_data_all"

dataset_ref = bigquery.DatasetReference(project, dataset_id)
```

```python
table_ref = dataset_ref.table("channel_stats")
table = client.get_table(table_ref)

channel_df = client.list_rows(table).to_dataframe()
```

```python
table_ref = dataset_ref.table("videos_stats")
table = client.get_table(table_ref)

videos_df = client.list_rows(table).to_dataframe()
```
